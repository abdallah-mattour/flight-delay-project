## 0. Bootstrap: install missing libraries automatically

The cell below checks for every library we depend on and installs the ones that are missing.
This makes the notebook fully self-bootstrapping: any team member (or the instructor) can run
it on a clean environment without first reading `requirements.txt`.

It uses `sys.executable -m pip` instead of a plain `!pip install` to guarantee the package
lands in the **same** Python interpreter that the notebook is running, which avoids the common
"I installed it but Jupyter still says it's missing" trap.

```python
# --- Auto-install any missing dependencies ---
# Safe to run multiple times: pip will simply skip packages that are already installed.

import importlib
import subprocess
import sys

# (import name -> pip package name).  Most match, but scikit-learn is the exception.
required = {
    "numpy":      "numpy",
    "pandas":     "pandas",
    "matplotlib": "matplotlib",
    "seaborn":    "seaborn",
    "sklearn":    "scikit-learn",
    "requests":   "requests",
}

missing = []
for import_name, pip_name in required.items():
    try:
        importlib.import_module(import_name)
    except ImportError:
        missing.append(pip_name)

if missing:
    print(f"Installing missing packages: {missing}")
    subprocess.check_call(
        [sys.executable, "-m", "pip", "install", "--quiet", *missing]
    )
    print("Installation complete. Please run all subsequent cells normally.")
else:
    print("All required libraries are already installed.")
```

## 4. Data Source Reachability Check

The project brief (Section 7) requires that the notebook must run from start to finish without
manual downloads. To honour this rule, every dataset we plan to use must be available at a
**stable public URL**. The cell below sends a lightweight request to each planned source to
confirm it is reachable. We deliberately do not download the files yet - that happens in
`01_data_collection.ipynb`.

**Planned sources for our 5 European countries (United Kingdom, Germany, France, Netherlands, Spain):**

| Source | URL | Used for |
|---|---|---|
| OurAirports - airports.csv | `https://davidmegginson.github.io/ourairports-data/airports.csv` | Airport metadata for all 5 countries |
| OurAirports - countries.csv | `https://davidmegginson.github.io/ourairports-data/countries.csv` | Country code lookup |
| OurAirports - runways.csv | `https://davidmegginson.github.io/ourairports-data/runways.csv` | Runway-level features |

The EUROCONTROL daily airport punctuality dataset is fetched in `01_data_collection.ipynb`
and covers all five of our European countries from a single source.

Note: some hosts reject requests that do not look like they come from a normal web browser.
We send a standard browser User-Agent header so the check reflects what a real download would do.

```python
import requests

sources = {
    "OurAirports - airports.csv":  "https://davidmegginson.github.io/ourairports-data/airports.csv",
    "OurAirports - countries.csv": "https://davidmegginson.github.io/ourairports-data/countries.csv",
    "OurAirports - runways.csv":   "https://davidmegginson.github.io/ourairports-data/runways.csv",
}

# A real browser-style User-Agent. Some static hosts (including GitHub Pages) refuse
# requests that come from default Python/requests user-agents, returning 403.
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
}

print(f"{'Source':<35} {'HTTP':<6} {'Status'}")
print("-" * 70)

for name, url in sources.items():
    try:
        # Ranged GET (first 1 KB only) instead of HEAD because some static
        # hosts do not respond meaningfully to HEAD requests.
        r = requests.get(url, headers={**HEADERS, "Range": "bytes=0-1023"}, timeout=20)
        code = r.status_code
        if code in (200, 206):
            status = "REACHABLE"
        elif code == 403:
            status = "BLOCKED (try again from a normal network)"
        else:
            status = "CHECK MANUALLY"
        r.close()
    except requests.RequestException as e:
        code = "ERR"
        status = f"NETWORK ERROR ({type(e).__name__})"
    print(f"{name:<35} {str(code):<6} {status}")
```
